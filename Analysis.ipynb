**A quintile** is a way to divide a set of data into five equal parts, where each part contains 20% of the total data points. This division helps in understanding the distribution of values within a dataset. For example, Q1 is the bottom 20% — the poorest fifth of households. Q5 is the top 20% — the richest fifth. Equal groups means exactly 20% of households fall in each bucket regardless of the actual naira amounts. Quintiles are often used in statistics to analyze income distribution, wealth, and other demographic data, providing insights into how different segments of a population are represented.

```python
# ── CHART 1: OOP Burden by Income Quintile ─────────────────────────────────
# Question: Do poorer households carry a heavier health spending burden?
# Expected: Q1 burden % should be significantly higher than Q5

quintile_burden = df_valid.groupby('income_quintile', observed=True).agg(
    median_burden   = ('oop_burden_pct', 'median'),
    mean_burden     = ('oop_burden_pct', 'mean'),
    household_count = ('oop_burden_pct', 'count')
).reset_index()

print("OOP Burden by Income Quintile:")
print(quintile_burden.to_string(index=False))
```

## Chart 3: The Geography of Health Poverty — OOP Burden by State

**Question:** Which Nigerian states carry the heaviest out-of-pocket health burden?

**Finding:** Kebbi state households spend 100% of recorded income on healthcare — 
the most extreme burden in the dataset. Kaduna follows at 55.6%. Seven of the 
top 10 highest-burden states are in the North, suggesting a regional 
concentration of health financing crisis that national policy must address 
specifically, not generically.

**Implication:** Northern state governments — particularly Kebbi, Kaduna, 
Adamawa, Jigawa, and Zamfara should be prioritised for Contributory Health 
Scheme rollout and federal NHIA subsidy allocation. FCT Abuja (0.7%) and 
Lagos (0.5%) sitting at the bottom confirms the urban economic advantage 
in health financing.

*Data note:* *Kebbi's 100% figure should be interpreted cautiously,
it may reflect a small sample of Kebbi households in the dataset rather 
than true median burden. Kaduna's 55.6% with a larger sample is a more 
reliable extreme case.*

## Chart 4: What Are Nigerian Households Actually Spending On?

**Question:** Where does healthcare money go — consultation fees, drugs, 
hospital stays, or transport?

**Finding:** Hospital stays (45.3%) and drug purchases (44.7%) together 
account for nearly 90% of all out-of-pocket health spending. Consultation 
fees (5.7%) and transport (3%) are minimal by comparison. Average hospital 
stay cost per person is ₦14,501 — more than 5x the average drug spend 
of ₦2,952.

**Implication:** Two targeted interventions would address 90% of the 
out-of-pocket burden: (1) Hospitalisation insurance covering catastrophic 
inpatient costs, and (2) subsidised essential medicines access for 
outpatient drug purchases. Consultation fee waivers alone — a common 
policy tool — would address less than 6% of the actual spending burden.

## Chart 5: How Many Households Are in Financial Crisis Right Now?

**Question:** What percentage of households in each income group are 
spending more than WHO's 10% catastrophic threshold on healthcare?

**Finding:** 80% of the poorest Nigerian households (Q1) are in 
catastrophic health expenditure — spending more than 10% of their 
income on healthcare. Even in Q3 (middle income), nearly half (46.8%) 
are above the catastrophic threshold. Only Q5 — the richest 20% — 
has a majority of households below the danger line, and even then 
8.7% are still in crisis.

**Implication:** Catastrophic health expenditure in Nigeria is not a 
fringe problem affecting a small minority. It is the normal experience 
for the bottom 60% of households by income. Universal health coverage 
expansion is not a welfare luxury — it is a financial emergency 
response for the majority of Nigerian households.
